```python
def func(r,x):
    return r*x-(x**3)
def f1(c,x):
    return (x**2)+c
def u(phi,x):
    return x*(1+x*phi)
```

```python
#This is Q5 on the homework
# range of r values to plot
rval= np.linspace(0,3.0,1000) # program has a runtime error when rmax is greater than 3
#program has an overflow error when rmin is less than 0
#1000 is the number of r steps
num=1000
x=np.ones(1000)
# Set the number of iterations to discard as "burn-in" 
burn_in = 100 
#loop over the range of r values and iterate the function map
for i in range(num):
    x=func(rval,x)
    if i > burn_in: # Just plotting enough to see the pattern so stopping when loop is at 101
        plt.plot(rval, x, ',k', alpha=0.05) #plotting all points r and x until i equals 101
        
plt.xlabel('r values')
plt.ylabel('x values')
plt.xlim(0,3)
plt.show()
```
